```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
newframe = pd.read_csv("label_flipping_experiments.csv")
```

```python
newframe.head()
```

```python
newframe.drop(columns=["Train_Loss", "Train_Acc", "Test_Loss", "Test_Acc", "Processing_Time", "Train_Loss.1", "Train_Acc.1", "Test_Loss.1", "Test_Acc.1", "Processing_Time.1"], axis=1, inplace=True)
```

```python
newframe.head()
```

```python
newframe['Model poisoning'].unique().tolist()
```

```python
xticks = newframe['poisoned nodes'].unique().tolist()
xticks

yticks = np.arange(0, 2.3, 0.2)
yticks
```

```python
newframe.pivot_table(index='poisoned nodes', columns='Model poisoning', values='Training Accuracy').plot(legend=True, xticks=xticks, title='Training Accuracy')
```

```python
newframe.pivot_table(index='poisoned nodes', columns='Model poisoning', values='Test Accuracy').plot(legend=True, xticks=xticks, title='Test Accuracy')
```

```python
newframe.pivot_table(index='poisoned nodes', columns='Model poisoning', values='Training Loss').plot(legend=True, xticks=xticks, title='Training Loss')
```

```python
newframe.pivot_table(index='poisoned nodes', columns='Model poisoning', values='Test Loss').plot(legend=True, xticks=xticks, title='Test Loss')
```

```python
newframe.pivot_table(index='poisoned nodes', columns='Model poisoning', values='Processing Time').plot(legend=True, xticks=xticks, title='Processing Time')
```

```python
fig, axs = plt.subplots(1, 2, figsize=(16,6))

newframe.pivot_table(index='poisoned nodes', columns='Model poisoning', values='Training Accuracy').plot(legend=True, xticks=xticks, title='Training Accuracy', ax=axs[0])
newframe.pivot_table(index='poisoned nodes', columns='Model poisoning', values='Test Accuracy').plot(legend=True, xticks=xticks, title='Test Accuracy', ax=axs[1])
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(16,6))

newframe.pivot_table(index='poisoned nodes', columns='Model poisoning', values='Training Loss').plot(legend=True, xticks=xticks,  yticks=yticks,title='Training Loss', ax=axs[0])
newframe.pivot_table(index='poisoned nodes', columns='Model poisoning', values='Test Loss').plot(legend=True, xticks=xticks, yticks=yticks, title='Test Loss', ax=axs[1])
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(16,6))

newframe.pivot_table(index='poisoned nodes', columns='Model poisoning', values='Training Loss').plot(legend=True, xticks=xticks, title='Training Loss', ax=axs[0])
newframe.pivot_table(index='poisoned nodes', columns='Model poisoning', values='Test Loss').plot(legend=True, xticks=xticks, title='Test Loss', ax=axs[1])
plt.show()
```
